```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

df.info()

df.head()

```

```python
new_df=df[((df['Age'] > 10)&(df['Embarked']=='S'))]
new_df.head()
```

```python
# Create a histogram for the 'age' column
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='age', kde=True, color='skyblue', bins=20)  # kde=True adds a Kernel Density Estimation plot
plt.title('Distribution of Age in Titanic Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

RESULTS:

```python
# Create a histogram for the 'survived' column
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='survived', kde=False, color='skyblue', bins=2)  # Since 'survived' is binary, we use 2 bins
plt.title('Survival Distribution in Titanic Dataset')
plt.xlabel('Survived (0: No, 1: Yes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a bar plot for survival based on gender
plt.figure(figsize=(10, 6))

sns.countplot(data=df, x='sex', hue='survived', palette='pastel')

# Set labels and title
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Add legend
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])

plt.show()
```

```python
# Filter data for men and women
men_data = df[df['sex'] == 'male'].head(20)
women_data = df[df['sex'] == 'female'].head(20)

# Create a dot plot
plt.figure(figsize=(10, 6))

sns.stripplot(x='sex', y='fare', data=men_data, jitter=True, color='blue', alpha=0.7, label='Men')
sns.stripplot(x='sex', y='fare', data=women_data, jitter=True, color='red', alpha=0.7, label='Women')

# Set labels and title
plt.title('Dot Plot of Ticket Fare for Men and Women')
plt.xlabel('Gender')
plt.ylabel('Ticket Fare')

# Add legend
plt.legend()

plt.show()
```
